```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Bengaluru_House_Data.csv")
df
```

```python
cols=["area_type","availability","society","balcony"]
df["society"].unique()
```

```python
df.drop(cols,axis=1,inplace=True)
```

```python
df.info()
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

```python
## cleaing location columns
df["location"].unique()
## removing spaces between the location
df["location"]=df["location"].str.strip()
df["location"].value_counts()
## handling nan values with top location
mode=df['location'].mode()[0]
df["location"]=df["location"].fillna(mode)
df["location"].isnull().sum()
## grouping 
```

```python
loc=df['location'].value_counts()
loc_less_10=loc[loc<=10]
df['location']=df['location'].apply(lambda x: "others" if x in loc_less_10 else x)
loc=df["location"].value_counts()
loc
```

```python
#cleaning size col
#handling nan values in size
df["size"]=df["size"].fillna("2 BHK")
#extract only the bhk number and remove unwanted chars
df["size"]=df["size"].apply(lambda x: int(str(x).split()[0]))
df["size"].unique()
#renaming size col to BHK
df.rename(columns={"size":"bhk"},inplace=True)
```

```python
#clean total_sqft col
def clean_total_sqft(sqft):
    try:
        l=sqft.split("-")
        if len(l)==2:
            return (float(l[0])+float(l[1]))/2
        else:
            return float(l[0])
    except:
        return None
df["total_sqft"]=df["total_sqft"].apply(clean_total_sqft)
```

```python
#handling nan values in bath
med=df["bath"].median()
df["bath"]=df["bath"].fillna(med)
df["bath"].unique()
```

```python
#create a new col price_per_sqft
#formula price_per_sqft = price/total_sqft
df["price_per_sqft"]=(df["price"]*100000)/df["total_sqft"]
```

```python
#handling outliers in total_sqft
#observation : in total_sqft col we have houses less than 300 sqft those are outliers 
#in real time houses starts with min of 300 sqft

#formula : total_sqft/bhk >=300 ----> it keeps all the realistic total sqft
df = df[df["total_sqft"]/df["bhk"] >= 300]
#total_sqft>=300
```

```python
df.describe()
```

```python
## handling outliers in BHK Column
df['bhk'].unique()
df=df[df['bhk']<=6]
```

```python
## handling outliers in bath column
df[df['bath']>df['bhk']+2][['bhk','bath']]
df=df[df['bath']<df['bhk']+2]
```

```python
## handling outliers in price_per_sqft
## IQR METHOD
q1=df["price_per_sqft"].quantile(0.25)
q3=df["price_per_sqft"].quantile(0.75)
IQR=q3-q1
lower=q1-0.5*IQR
upper=q3+0.5*IQR
df=df[(df['price_per_sqft']>=lower) & (df['price_per_sqft']<=upper)]
sns.boxplot(x="price_per_sqft",data=df)
```

```python
## remove pp sfqt
df=df.drop("price_per_sqft",axis=1)
df
```

```python
df.reset_index(inplace=True)
df=df.drop("index",axis=1)
df
```

```python
plt.scatter(df['total_sqft'],df['price'])
plt.xlabel("clean_total_sqft")

plt.ylabel("price")
plt.title("total sqft vs price")
```

```python
corr_data=df.iloc[:,1:].corr()
sns.heatmap(corr_data,annot=True)
```

```python
## loc vs avg price
df_new=df.groupby("location")["price"].mean().sort_values()
df_new
```

```python
## model training
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
## encoding loc column
en=LabelEncoder()
df["en_loc"]=en.fit_transform(df["location"])
df
```

```python
x=df.drop({'location','price'},axis=1)
y=df["price"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
print("traing eff:",model.score(xtrain,ytrain))
print("test eff:",model.score(xtest,ytest))
print("r2:",r2_score(ytest,ypred))
print("mae:",mean_absolute_error(ytest,ypred))
```

```python
dt=DecisionTreeRegressor()
dt.fit(xtrain,ytrain)
```

```python
ypred=dt.predict(xtest)
ypred
```

```python
print("traing eff:",model.score(xtrain,ytrain))
print("test eff:",model.score(xtest,ytest))
print("r2:",r2_score(ytest,ypred))
print("mae:",mean_absolute_error(ytest,ypred))
```
